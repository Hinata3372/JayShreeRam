```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return (x + 3)**2
```

```python
def df(x):
    return 2*x + 6
```

```python
def gradient_descent(initial_x, learning_rate, no_of_iterations):
    x = initial_x
    x_history = [x]

    for i in range (no_of_iterations):
        gradient = df(x)
        x = x - learning_rate * gradient
        x_history.append(x)

    return x, x_history
```

```python
initial_x = 2
learning_rate = 0.1
no_of_iterations = 50
```

```python
x, x_history = gradient_descent(initial_x, learning_rate, no_of_iterations)
```

```python
print("Local minimum {:.2f}".format(x))
```

```python
for i in range(no_of_iterations):
    print(f"Iteration ", (i+1), " : ",x_history[i])
```

```python
x_vals = np.linspace(-11,5,50)
```

```python
plt.plot(x_vals, f(x_vals))
plt.plot(x_history, f(np.array(x_history)), 'rx')
plt.show
```
